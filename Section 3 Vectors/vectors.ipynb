###      VIDEO: Vector multiplication: Dot product 


```python
## many ways to compute the dot product

v1 = np.array([ 1, 2, 3, 4, 5 ])
v2 = np.array([ 0, -4, -3, 6, 5 ])

#method 1 
dp1=sum ( np.multiply(v1,v2)) #giong nhu phep cong tat ca cac tich lai 

#method 2 
dp2=np.dot(v1,v2)

#method 3
dp3=np.matmul(v1,v2)

#method 4
dp4=0 
for i in range(0, len(v1)):
        dp4 = dp4 + v1[i]*v2[i]

print(dp1, dp2, dp3, dp4)
```

###      VIDEO: Dot product properties: associative and distributive  


```python
## distributive property 

#create random vectors 
n=10
a=np.random.randn(n)
b=np.random.randn(n)
c=np.random.randn(n)

#the two results
res1=np.dot(a, (b+c))
res2=np.dot(a,b) + np.dot(a,c)

#print and compare 
print(res1, res2)
```

```python
## distributive property 

#create random vectors 
n=5
a=np.random.randn(n)
b=np.random.randn(n)
c=np.random.randn(n)

#the two results
res1= np.dot(a, np.dot(b,c))
res2= np.dot(np.dot(a,b),c)

#print and compare 
print(res1)
print(res2)
```
